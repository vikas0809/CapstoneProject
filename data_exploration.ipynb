```python
# Import dependencies 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load a CSV file into DataFrame
claims_df = pd.read_csv('train_data.csv')
```

```python
# number of rows and columns in the data frame
print("Shape:",claims_df.shape)
```

```python
# View the Target Column "Is_Claim"
Is_claim = claims_df["is_claim"].tolist()
print("Number of 0's:", Is_claim.count(0))
print ("Number of 1's:", Is_claim.count(1)) 
```

```python
print ("Number of 1's:" Is_claim.count(1)) 
```

```python
# View the first 5 rows of the data frame
print(claims_df.head())
```

```python
# Basic statistics about the data
print(claims_df.describe())
```

```python
# Get information about the columns 
print(claims_df.info())
```

```python
# Plot histograms for all numerical columns 
claims_df.hist(bins=50, figsize=(20,15))
plt.show()
```

```python
# Visualize the relationship between all numerical columns 
pd.plotting.scatter_matrix(claims_df, figsize=(12,8))
plt.show()
```

```python
# Get correlation matrix
corr = claims_df.corr()
print("\nCorrelation Matrix:")
print(corr)
```

```python
# Plot boxplots for all numeric columns
claims_df.select_dtypes(include=[np.number]).plot(kind='box', subplots=True, layout=(18,18), sharex=False, sharey=False, figsize=(30,15))
plt.show()
```
